```python
import numpy as np
import pandas as pd

# Load the dataset
file_path = "/content/Gold_ demand_tonnes_yearly.csv"
df = pd.read_csv(file_path)

gold_demand_2023 = df["2023"].values
mean_2023 = np.mean(gold_demand_2023)
std_2023 = np.std(gold_demand_2023)

print("Mean gold demand in 2023:", mean_2023)
print("Standard deviation of gold demand in 2023:", std_2023)
```

```python
median_2023 = np.median(gold_demand_2023)
variance_2023 = np.var(gold_demand_2023)

print("Median gold demand in 2023:", median_2023)
print("Variance of gold demand in 2023:", variance_2023)
```

```python
#Find the Country with the Highest Gold Demand in 2023
max_demand_country = df.loc[df["2023"].idxmax(), "Countries"]
print("Country with the highest gold demand in 2023:", max_demand_country)
```

```python
#Find Countries Where Gold Demand in 2023 is Above the Mean
above_mean_countries = df[df["2023"] > mean_2023]["Countries"].tolist()
print("Countries with demand above the mean in 2023:", above_mean_countries)
```

```python
#Sort the Data by Average Demand
sorted_df = df.sort_values(by="Average demand", ascending=False)
print("Top 5 countries by average demand:\n", sorted_df.head())
```

```python
# Find the Top 3 Countries with the Highest Year-on-Year Percentage Change
top_3_growth = df.nlargest(3, "Year-on-year % change")
print("Top 3 countries with highest year-on-year % change:\n", top_3_growth)
```

```python
#Filter Countries with Positive Year-on-Year Percentage Change

positive_growth_df = df[df["Year-on-year % change"] > 0]
print("Countries with positive year-on-year % change:\n", positive_growth_df)
```

```python
print(df.info())
print(df.describe())
```

```python
#Bar Chart for Top 5 Countries by Gold Demand in 2023
import matplotlib.pyplot as plt

top_5_countries = sorted_df.head(5)

plt.figure(figsize=(10, 5))
plt.bar(top_5_countries["Countries"], top_5_countries["2023"], color='gold')
plt.xlabel("Countries")
plt.ylabel("Gold Demand in 2023 (tonnes)")
plt.title("Top 5 Countries by Gold Demand in 2023")
plt.xticks(rotation=45)
plt.show()
```

```python
#Histogram: Distribution of Gold Demand in 2023
plt.figure(figsize=(8, 5))
plt.hist(df["2023"], bins=10, color='blue', edgecolor='black', alpha=0.7)
plt.xlabel("Gold Demand in 2023 (tonnes)")
plt.ylabel("Frequency")
plt.title("Distribution of Gold Demand in 2023")
plt.grid()
plt.show()
```

```python
# Pie Chart: Share of Gold Demand for Top 5 Countries in 2023

top_5 = df.nlargest(5, "2023")  # Get top 5 countries

plt.figure(figsize=(8, 5))
plt.pie(top_5["2023"], labels=top_5["Countries"], autopct='%1.1f%%', colors=['gold', 'blue', 'red', 'green', 'purple'])
plt.title("Gold Demand Share (Top 5 Countries in 2023)")
plt.show()
```
